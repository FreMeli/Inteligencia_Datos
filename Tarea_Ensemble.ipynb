```python
import numpy as np

# Función para realizar un experimento de lanzamiento de una moneda sesgada
def lanzar_moneda(prob_cara, num_lanzamientos):
    resultados = np.random.choice([1, 0], size=num_lanzamientos, p=[prob_cara, 1 - prob_cara])
    return resultados

# a) Simular 1000 lanzamientos con una probabilidad de cara 0.51
np.random.seed(42)  # Para hacer el experimento reproducible
lanzamientos = lanzar_moneda(0.51, 1000)
mayoria_caras = np.sum(lanzamientos)
print(f"Total de caras en 1000 lanzamientos: {mayoria_caras}")
```

## Parte 2

image.png

```python
# b) Repetir el experimento 10,000 veces para 1000 lanzamientos y determinar la probabilidad de mayoría de caras
num_experimentos = 10000
mayorias = []
for _ in range(num_experimentos):
    lanzamientos = lanzar_moneda(0.51, 1000)
    mayorias.append(np.sum(lanzamientos) > 500)  # Mayoría de caras es cuando > 500

probabilidad_mayoria = np.mean(mayorias)
print(f"Probabilidad de obtener una mayoría de caras en 1000 lanzamientos: {probabilidad_mayoria * 100:.2f}%")
```

##Parte 3

image.png

```python
import matplotlib.pyplot as plt

# c) Repetir el experimento aumentando el número de lanzamientos de 500 a 10,000 en incrementos de 10
lanzamientos_incrementos = np.arange(500, 10001, 10)
probabilidades_mayoria = []

for num_lanzamientos in lanzamientos_incrementos:
    mayorias = []
    for _ in range(num_experimentos):
        lanzamientos = lanzar_moneda(0.51, num_lanzamientos)
        mayorias.append(np.sum(lanzamientos) > (num_lanzamientos / 2))  # Mayoría de caras
    probabilidades_mayoria.append(np.mean(mayorias))

# Graficar la probabilidad de obtener una mayoría de caras vs número de lanzamientos
plt.plot(lanzamientos_incrementos, probabilidades_mayoria)
plt.xlabel('Número de lanzamientos')
plt.ylabel('Probabilidad de mayoría de caras')
plt.title('Probabilidad de obtener una mayoría de caras vs número de lanzamientos')
plt.grid(True)
plt.show()
```
